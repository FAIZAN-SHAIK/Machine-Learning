```python
import pandas as pd
```

```python
df = pd.read_csv("deliveries.csv")
df.head(50)
```

### top 5 batsman in terms of runs scored

```python
players = df.groupby('batsman')
```

```python
players.first().shape
```

```python
players['batsman_runs'].sum()
```

```python
players['batsman_runs'].sum().sort_values(ascending=False).head()
```

### to findout the batsman who hit more 4's or 6's|

```python
mask = df['batsman_runs']==6
f_king = df[mask]
f_king
```

#### no:of 4's till date

```python
f_king.shape[0]
```

```python
f_king.groupby('batsman').size().sort_values(ascending=False).head()
```

```python
f_king.groupby('batsman')['batsman_runs'].count().sort_values(ascending=False).head()
```
